```python
import pandas as pd
import bokeh as bk
import seaborn as sb
%matplotlib inline
from bokeh.plotting import *
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.models.glyphs import Line as Line_glyph, Oval
from bokeh.models import Range1d, LinearAxis, FactorRange, LabelSet
from bokeh.charts import Bar, Scatter, BoxPlot
from bokeh.charts.attributes import cat
from bokeh.layouts import gridplot
import numpy as np
pd.options.display.max_rows = 50
pd.options.display.max_columns = 999
```

```python
mozilla_df = pd.read_csv("Mozilla_Data.csv", encoding = "iso-8859_1", low_memory = False)
mozilla_df = mozilla_df.fillna(0)
mozilla_df
```

```python
skinny_df = mozilla_df[['Response ID','WiFi Router:Check all the internet connected devices you currently own:','Laptop computer:Check all the internet connected devices you currently own:', 'Smart phone:Check all the internet connected devices you currently own:','Smart TV:Check all the internet connected devices you currently own:','Activity Tracker (ex: Fitbit or Apple Watch):Check all the internet connected devices you currently own:','Smarthome Hub (ex. Amazon Echo, Google Alexa):Check all the internet connected devices you currently own:','Car that connects to the internet:Check all the internet connected devices you currently own:','Smart Thermostat (ex: Nest):Check all the internet connected devices you currently own:','Smart Appliance (ex. Coffeemaker, Refrigerator, Oven, Fridge):Check all the internet connected devices you currently own:','Smart Lighting (ex. Connected lighting switches, dimmers, or bulbs):Check all the internet connected devices you currently own:']].copy()
skinny_df
```

```python
#Finding the amount of devices owned per group
count_df = pd.get_dummies(skinny_df)
count_df.columns = ['Response ID','2','WiFi','4','Laptop','6','SmartPhone','8','SmartTV','10','Activity Tracker','12','SmartHome','14','Car','16','Smart Thermostat','18','Smart Appliance','20','Smart Lighting',]
count_df = count_df[['Response ID','WiFi','Laptop','SmartPhone','SmartTV','Activity Tracker','Smart Thermostat','Smart Appliance','Smart Lighting']].copy()
count_df['Total Devices Owned'] = count_df.sum(axis = 1).subtract(count_df['Response ID'])
count_df = count_df[['Response ID', 'Total Devices Owned']]
count_df.head(10)
```

```python
#Merging the Devices Owned per Response ID
graph_df = pd.merge(mozilla_df, count_df, on = 'Response ID')
graph_df.head()
```

```python
#Calculating Sums of Devices owned per group
g1 = graph_df[graph_df['What is your biggest fear as we move towards a more connected future?'] == 'The loss of privacy']
Privacy_Total = g1['Total Devices Owned'].sum()
g2 = graph_df[graph_df['What is your biggest fear as we move towards a more connected future?'] == 'WeÛªll lose touch with one another']
Lose_Touch = g2['Total Devices Owned'].sum()
g3 = graph_df[graph_df['What is your biggest fear as we move towards a more connected future?'] == 'WeÛªll be less safe']
Less_Safe = g3['Total Devices Owned'].sum()
g4 = graph_df[graph_df['What is your biggest fear as we move towards a more connected future?'] == 'Other (please specify)']
Other = g4['Total Devices Owned'].sum()
g5 = graph_df[graph_df['What is your biggest fear as we move towards a more connected future?'] == 'I have no fears about a more connected future']
No_Fear = g5['Total Devices Owned'].sum()
```

```python
#Throwing into Dict for Easy Readability
total_devices_owned = dict()
total_devices_owned['Privacy Total'] = Privacy_Total
total_devices_owned['Lose Touch'] = Lose_Touch
total_devices_owned['Less Safe'] = Less_Safe
total_devices_owned['Other'] = Other
total_devices_owned['No Fear'] = No_Fear
total_devices_owned
```
